```python
matrix = []
for i in range(len(df['Close'])-1, 2, -1):
  
  day3=df['Close'][i-3]
  day2=df['Close'][i-2]
  day1=df['Close'][i-1]
  matrix.append([df['Close'][i], day1, day2, day3])

matrix
```

```python
data_lstm=pd.DataFrame(matrix,columns=['Close','One day ago','Two days ago','Three days ago'])
data_lstm
```

```python
from sklearn.model_selection import train_test_split

X = data_lstm.drop(['Close'], axis = 1)
Y = data_lstm['Close'].values

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
X_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
import numpy as np 
```

```python
model = Sequential()

model.add(LSTM(100, input_shape=(1,3), activation='relu', return_sequences=True))

model.add(LSTM(64, activation='relu'))

model.add(Dense(1 ,activation='linear'))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])
     

history = model.fit(X_train, Y_train, shuffle=False, validation_split=0.2, epochs=200, batch_size=16)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2_score(Y_test,y_pred)
```

```python
from transformers import AutoTokenizer, XLMForSequenceClassification,RobertaForSequenceClassification
```

```python
tret_data=df.drop(['Top'+str(i) for i in range(1,11)],axis=1)
```

```python
tret_data=tret_data.drop(['Label'],axis=1)
tret_data
```

```python
from sklearn.model_selection import train_test_split
X=tret_data.drop(['Close'],axis=1)
Y=tret_data.Close
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
df_train = pd.concat([X_train,Y_train],axis=1)
df_test= pd.concat([X_test,Y_test],axis=1)
df_train
```

```python
df_train.columns = ['Open','combined_news','label']
df_test.columns = ['Open','combined_news','label']
```

```python
df_train.to_csv('train.csv',index=False)
df_test.to_csv('test.csv',index=False)
```

```python
dataset = load_dataset("csv", data_files={"train": "train.csv", "test": "test.csv"})
```

```python
encoded_dataset = dataset.map(lambda t: tokenizer(t['combined_news'],  truncation=True), batched=True,load_from_cache_file=False)
```

```python
encoded_dataset
```

```python
from transformers import TrainingArguments, Trainer
arg = TrainingArguments(
    "sentiment",
    learning_rate=5e-5,
    num_train_epochs=1,
    per_device_eval_batch_size=8,
    per_device_train_batch_size=8,
    seed=19
)
trainer = Trainer(
    model=model,
    args=arg,
    tokenizer=tokenizer,
    train_dataset=encoded_dataset['train'],
    eval_dataset=encoded_dataset['test']
)
```

```python
trainer.train()
```

```python
y_pred = trainer.predict(encoded_dataset['test'])
```

```python
y_pred.predictions
```

```python
y_pred.metrics
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred.predictions,Y_test)
```
