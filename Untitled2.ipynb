```python
message_embeddings = embed(df.OriginalTweet.values)
```

```python
from sklearn.decomposition import PCA
import numpy as np

data = message_embeddings
pca = PCA(150)
 
df_new = pca.fit_transform(data)
```

```python
df_new=pd.DataFrame(df_new)
```

```python
df_new
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, Y_train, Y_test = train_test_split(df_new, df.Sentiment, test_size = 0.2, random_state = 0)
     
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
```

```python
df.Sentiment.unique()
```

```python
Y_train_cat=keras.utils.to_categorical(Y_train,num_classes=5)
Y_test_cat=keras.utils.to_categorical(Y_test,num_classes=5)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout

model = Sequential()

model.add(Dense(64, input_dim=150, kernel_initializer='uniform', activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(32, kernel_initializer='uniform', activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(16, kernel_initializer='uniform', activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform', activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) 
```

```python
model.fit(X_train, Y_train, epochs=150, batch_size=10) 
```

```python
y_pred = model.predict(X_test) 
```

```python
import numpy as np
```

```python
pred=np.argmax(y_pred,axis=1)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

cm = confusion_matrix(Y_test, pred)
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d", cbar=False)

ac = accuracy_score(Y_test, pred)
print('accuracy of the model: ', ac)
```

```python
print(f1_score(Y_test,pred,average='macro'))
```
